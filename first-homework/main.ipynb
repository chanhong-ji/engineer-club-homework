```python
import requests
import openai
from typing import List, Dict
```

```python
base_url = "https://nomad-movies.nomadcoders.workers.dev"


def get_popular_movies() -> List[Dict]:
    response = requests.get(f"{base_url}/movies")
    return response.json()


def get_movie_details(movie_id: int) -> Dict:
    response = requests.get(f"{base_url}/movies/{movie_id}")
    return response.json()


def get_movie_credits(movie_id: int) -> Dict:
    response = requests.get(f"{base_url}/movies/{movie_id}/credits")
    return response.json()


def create_prompt(question: str) -> str:
    return f"""
    너는 아래 함수 중 정확히 하나의 함수 호출만 출력하는 거야.

    사용 가능한 함수 목록:
    - get_popular_movies() -> List[Dict]
    - get_movie_details(movie_id: int) -> Dict
    - get_movie_credits(movie_id: int) -> Dict

    규칙:
    - 반드시 위 3개 중 하나의 함수명만 사용한다. 
    - 부가 설명 없이 함수 호출 한 줄만 출력한다.

    질문:
    {question}
    """


questions = [
    "지금 인기 있는 영화가 무엇인지 알려줘",
    "movie ID 550에 해당하는 영화가 무엇인지 알려줘",
    "movie ID 550에 해당하는 영화에 누가 출연하는지 알려줘",
]
```

```python
client = openai.OpenAI()

for question in questions:
    response = client.chat.completions.create(
        model="gpt-4o-mini",
        messages=[
            {"role": "system", "content": create_prompt(question)},
        ],
    )
    print(response.choices[0].message.content)
```
