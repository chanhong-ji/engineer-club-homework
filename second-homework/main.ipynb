```python
import openai
from typing import List, Dict
```

```python
client = openai.OpenAI()
```

```python
def call_ai_with(messages: List[Dict]):
    response = client.responses.create(
        model="gpt-5.4-mini",
        input=messages,
    )
    messages += [{"role": el.role, "content": el.content} for el in response.output]
    print(response.output_text)
    return messages
```

```python
messages = []


while True:
    new_message_text = input("You: ")
    if new_message_text == "quit" or new_message_text == "q":
        break
    else:
        messages.append({"role": "user", "content": new_message_text})

    messages = call_ai_with(messages)
```
